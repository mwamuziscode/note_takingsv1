```python
import turtle
import math

def draw_fibonacci_spiral(n):
    screen = turtle.Screen()
    screen.bgcolor("black")
    
    fib_turtle = turtle.Turtle()
    fib_turtle.speed(0)
    fib_turtle.pensize(2)
    fib_turtle.pencolor("gold")
    
    a, b = 0, 1
    for _ in range(n):
        fib_turtle.circle(b, 90)
        a, b = b, a + b
    
    fib_turtle.hideturtle()
    screen.mainloop()

# Change the number to increase/decrease Fibonacci levels
draw_fibonacci_spiral(10)
```

```python
import turtle

def draw_box(x, y, width, height):
    """Draws a rectangle with top-left corner at (x, y)."""
    turtle.penup()
    turtle.goto(x, y)
    turtle.pendown()
    for _ in range(2):
        turtle.forward(width)
        turtle.right(90)
        turtle.forward(height)
        turtle.right(90)

def draw_hadamard_gate(x, y):
    """Draws a Hadamard gate as a box labeled 'H'."""
    draw_box(x, y, 40, 40)
    turtle.penup()
    turtle.goto(x + 15, y - 25)
    turtle.pendown()
    turtle.write("H", font=("Arial", 16, "bold"))

def draw_beam_splitter(x, y):
    """Draws a diagonal line representing a beam splitter."""
    turtle.penup()
    turtle.goto(x, y)
    turtle.pendown()
    turtle.goto(x + 30, y - 30)

def draw_path(points, dashed=False):
    """Draws a path connecting given points, optionally dashed."""
    turtle.penup()
    for i, (x, y) in enumerate(points):
        if dashed and i % 2 == 1:
            turtle.penup()
        else:
            turtle.pendown()
        turtle.goto(x, y)

def main():
    turtle.speed(3)
    
    # Draw outer box
    draw_box(-150, 100, 300, -200)
    
    # Draw input and output labels
    turtle.penup()
    turtle.goto(-170, 50)
    turtle.pendown()
    turtle.write("In", font=("Arial", 16, "bold"))

    turtle.penup()
    turtle.goto(170, 50)
    turtle.pendown()
    turtle.write("Out", font=("Arial", 16, "bold"))

    # Draw Hadamard gate
    draw_hadamard_gate(-100, 50)

    # Draw paths
    draw_path([(-170, 50), (-100, 50)])  # Input to Hadamard gate
    draw_path([(-60, 50), (0, 50)], dashed=True)  # Hadamard to splitter
    draw_beam_splitter(0, 50)
    draw_path([(30, 20), (100, 20)], dashed=True)  # Path after splitter

    # Draw detector (square)
    draw_box(100, 30, 40, -40)

    # Draw final output arrow
    draw_path([(140, 10), (170, 10)]) 

    turtle.done()

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk

class QuantumCircuitApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Quantum Circuit Simulation")

        # Canvas for drawing the quantum circuit
        self.canvas = tk.Canvas(root, width=400, height=250, bg="white")
        self.canvas.pack()

        # Draw circuit components
        self.draw_circuit()

        # Create button to move the electron
        self.button = tk.Button(root, text="Send Electron", command=self.animate_electron)
        self.button.pack(pady=10)

        # Create electron as a small circle
        self.electron = self.canvas.create_oval(50, 90, 70, 110, fill="red")

    def draw_circuit(self):
        """Draws the quantum circuit components on the canvas."""
        # Outer box
        self.canvas.create_rectangle(30, 50, 370, 200, outline="black")

        # Labels
        self.canvas.create_text(20, 100, text="In (Electron)", font=("Arial", 12, "bold"))
        self.canvas.create_text(380, 100, text="Out", font=("Arial", 12, "bold"))

        # Hadamard Gate (H)
        self.canvas.create_rectangle(100, 80, 140, 120, outline="black", fill="lightgray")
        self.canvas.create_text(120, 100, text="H", font=("Arial", 16, "bold"))

        # Beam Splitter (diagonal line)
        self.canvas.create_line(200, 70, 230, 100, width=3)

        # Detector (final output)
        self.canvas.create_rectangle(300, 80, 340, 120, outline="black", fill="gray")

        # Electron path (dashed lines)
        self.dashed_line(70, 100, 100, 100)  # Input to Hadamard
        self.dashed_line(140, 100, 180, 100)  # Hadamard to splitter
        self.dashed_line(230, 100, 300, 100)  # Splitter to output

    def dashed_line(self, x1, y1, x2, y2, dash_length=5):
        """Draws a dashed line between two points."""
        dx = x2 - x1
        dy = y2 - y1
        steps = int(((dx**2 + dy**2) ** 0.5) / (2 * dash_length))

        for i in range(steps):
            start_x = x1 + (dx * i / steps)
            start_y = y1 + (dy * i / steps)
            end_x = x1 + (dx * (i + 0.5) / steps)
            end_y = y1 + (dy * (i + 0.5) / steps)
            self.canvas.create_line(start_x, start_y, end_x, end_y, fill="black")

    def animate_electron(self):
        """Moves the electron through the quantum circuit step by step."""
        positions = [(90, 100), (160, 100), (210, 100), (270, 100), (320, 100)]
        
        for i, (x, y) in enumerate(positions):
            self.root.after(i * 300, self.move_electron, x, y)  # Delay each step

    def move_electron(self, x, y):
        """Moves the electron (red circle) to a new position."""
        self.canvas.coords(self.electron, x-10, y-10, x+10, y+10)

# Run the Tkinter app
if __name__ == "__main__":
    root = tk.Tk()
    app = QuantumCircuitApp(root)
    root.mainloop()
```
